```python
from pypot.feetech import FeetechSTS3215IO
import time
```

```python
gripper = FeetechSTS3215IO(port='/dev/LeRobotFollower', baudrate="1000000")
```

```python
ID = 6
```

```python
gripper.enable_torque([ID])
```

```python
gripper.disable_torque([ID])
```

```python
gripper.get_present_position([ID])
```

```python
gripper.get_present_position([ID])
```

```python
# PincOpen
gripper.set_lock({ID: 0})
time.sleep(0.5)
gripper.set_acceleration({ID: 200})
time.sleep(0.5)

gripper.set_max_angle_limit({ID: 0})
time.sleep(0.5)
gripper.set_min_angle_limit({ID: -147})
time.sleep(0.5)

# Set the maximum output torque limit of the servo, and set 1000 = 100% * locked torque,Power on assigned to address 48 torque limit“
gripper.set_torque_limit({ID: 1000})
time.sleep(0.5)

# The maximum torque threshold of starting overload protection time meter, if set to 80, means 80% of the maximum torque
# Default values is 40 in PincOpen
gripper.set_overload_torque({ID: 30})
time.sleep(0.5)

# After entering the overload protection, the output torque, if set to 20, means 20% of the maximum torque
gripper.set_protective_torque({ID: 5})
time.sleep(0.5)

# The timing time when the current load output exceeds the overload torque and remains. If 200 is set to 2 seconds, the maximum can be set to 2.5 seconds
gripper.set_protection_time({ID: 7})
time.sleep(0.5)

# Custom
# gripper.set_maximum_acceleration({ID: 254})
# time.sleep(0.5)

gripper.set_lock({ID: 1})
```

```python
def open():
    gripper.enable_torque([ID])
    gripper.set_goal_position({ID: -140})

def close():
    gripper.enable_torque([ID])
    gripper.set_goal_position({ID: 0})
```

```python
from ipywidgets import interact

@interact(x=(-10, 90, 5))
def show_value(x=0):
    gripper.enable_torque([ID])
    gripper.set_goal_position({ID: x})
```

```python
from ipywidgets import interact

@interact(toggle=widgets.ToggleButton(description='Open', value=False))
def toggle_action(toggle):
    if toggle:
        print('State: Closed')
        close()
    else:
        print('State: Open')
        open()
```
